```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# **Converting censius to fahrenheit**

Data finding and preparation

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

Choosing a model

```python
model = tf.keras.Sequential(layers=tf.keras.layers.Dense(units=1, input_shape=[1]))
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss=tf.keras.losses.mean_squared_error)

```

Training model

```python
history = model.fit(celsius, fahrenheit, epochs=500, verbose=False)
```

Data visualization

```python
plt.plot(history.history["loss"])
```

Predict

```python
print(model.predict([38]))
```
